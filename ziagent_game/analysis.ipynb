## Setup and Description
### Game setup
For this assignment, I chose to look at the game of chicken. In the game of chicken each player can choose to swerve or stay straight. Their payoff depends on the decision of the other player. As noted in class, this game does not have a dominant strategy, however there are payoffs depending on what action each player takes.

Breakdown of payoffs

- P1 swerves, P2 swerves -- both get 1
- P1 swerves, P2 stays straight -- P1 gets 0, P2 gets 2
- P1 stays straight, P2 swerves -- P1 gets 2, P2 gets 0
- P1 stays straight, P2 stays straight -- both get 0

### Special setup of agents:
For further analysis, extra functionality was added to the code beyond what is needed to establish the game of chicken. Agents have added functionality of memory. Memory refers to how many games they remember and how good the payoff was for those games. When memory is disabled, an agent randomly choose a strategy. When memory is enabled, the first few turns will be random. The agent then creates a strategy based off of the size of memory. The agent will use memory to help them decide what their next move should be. If agents have a memory of three games, then they take their moves and payoffs from the last three games into account.

Added to the decision process is also a check for agents to see if they are stuck in a rut. If an agent is stuck in rut, then memory is forgotten and a random move is made. The length of time considered for a rut is two times memory. If memory is three games, then the last six moves are considered for rut analysis. If all moves are the same, then the agent picks a random move.

## Analysis

In the first part of experient, I looked at runs of 100 agents playing 100 games and varied the amount of memory in multiple runs. 

For each of the runs, the outcome is consistent when running the same settings.When there is no memory, agents stabilize to a steady-state with an equal chance of swerving or staying straight. When agents are set to remember the last game, and to take that into consider, the population stabilizes with about 1.5 times the tendency to swerve over staying straight. However, when the memory is set to remember the last three games, a group of agents break away from the ‘norm’  and create a sub-group that strategizes by moving straight more often than swerving. In effect, they take advantage of the possibility of extra profits, since the chances are they will be in a game with another agent that has a swerving strategy. When the agents have a longer term memory of five or more games, there is a clear delineation, between strategies that agents adopted. Agents either swerve or stay straight.

With respect to total payoffs, as memory increases the payoffs move from a normal distribution to one that is skewed to the right. This is true until about the agent memory of five games, when the agents total payoff distribution reaches a peak payoff. When we reach a memory of 10 or 20 games, the total playoff distribution decreases.

When taking a look at the number of games and the speed of convergence, it appears that stabilization of a steady occurs more quickly when the memory attribute is present and with larger populations. With respect to populations, the density of a norm becomes more apparent more quickly. (See figure in ‘Memory of 3 with Population change’ subsection.)

As the model stabilizes, agents arrive at a Nash equilibria when using a strategy of memory recollection. It can also be said that agent-level equilibria emerge. This occurs around the point when agents have a memory of five games. Social norms arise during this process. The more memory an agent has the more the agents become bifurcated in their strategies. A social norm of swerving is dominant, while a secondary social norm of straight also becomes clear. Both norms emerge quickly.
