### Setup:
For this assignment, I chose to look at the game of chicken. In the game of chicken each player can choose to swerve or stay straight. Their payoff depends on the decision of the other player. As noted in class, this game does not have a dominant strategy, however there are payoffs depending on what action each player takes.

Breakdown of payoffs

- P1 swerves, P2 swerves -- both get 1
- P1 swerves, P2 stays straight -- P1 gets 0, P2 gets 2
- P1 stays straight, P2 swerves -- P1 gets 2, P2 gets 0
- P1 stays straight, P2 stays straight -- both get 0

For further analysis, extra functionality was added to the code. Agents have added functionality of memory. Memory refers to how many games they remember and how good the payoff was for those games. When memory is enabled, an agent randomly choose a strategy on the amount of turns that exist less than memory. Then memory kicks in, and the agent will use memory to help them decide what their next move should be. 

Added to the decision process is also a check for agents to see if they are stuck in a rut. If an agent is stuck in rut, then memory is forgotten and a random move is made. The length of time considered for a rut is two times memory. If memory is three games, then the last six moves are considered for rut analysis. If all moves are the same, then the agent picks a random move.
