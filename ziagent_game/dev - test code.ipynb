```python
import numpy as np

from ziagent_game import play_game
from ziagent_game.agents import Agent
from ziagent_game.graphs import graph_moves, graph_payoffs


def run_and_graph(p, r, m, g_type):
    ''' Function to run games & analysis for all game variations.'''
    play_game.run(players, rounds, memory)

    # Create a list of agents to evaluate
    agents = Agent.instances
    swerves = []
    straights = []
    payoffs = []
    for a in agents:
        swerves.append(a[2])
        straights.append(a[3])
        payoffs.append(a[4])
    
    print "Figure A: %s" % g_type
    print "Swerve Mean: %s" % round(np.mean(swerves), 1)
    print "Swerve Std: %s" % round(np.std(swerves), 2)
    print "Straight Mean: %s" % round(np.mean(straights), 1)
    print "Straight Std: %s" % round(np.std(straights), 2)
    graph_moves(swerves, straights)
    
    print "Figure B: %s" % g_type
    print "Mean: %s" % round(np.mean(payoffs), 1)
    print "Std: %s" % round(np.std(payoffs), 2)
    graph_payoffs(payoffs, rounds)
```

```python
players = 10
rounds = 10
memory = 0
g_type = 'Control Group'
run_and_graph(players, rounds, memory, g_type)
```
