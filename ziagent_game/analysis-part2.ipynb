## Setup and Description
### Game setup
For this assignment, I chose to play with noise in the game of chicken. It should be noted that noise was added in the previous assignment, but the effect of the noise was not documented. Also added to this version of the game are time tick tracking.

### Special setup of agents:
To expand the previous set up of agents, I expanded the funtionality of whether or not an agent identifies themselves as being in a 'rut'. A rut is defined as the precentage of dominate moves being greater than the rut_percentage. So, if the rut_percentage is 75% and the agent has executed at least 75% of one move type (swerve or straight), then the agent is in a rut. At that point the agent selects a random move. In the previous game, a rut was defined as 100% of moves in memory being equal. 

The rut check does not go into effect until at least twice the memory of games have been played. This is to give the agent the ability to explore on their own, before noise comes into effect. The rut check also considers twice the memory when giving consideration to whether or not the agent is in a rut. The purpose of this is to see if the previous set of memory set the tone for the current set of memory.

It should be noted that a rut percentage of 50% in effect would counteract memory, because it would force the agent to be random. A rut percentage of less than 50% would have no effect, because the rut manager looks to see if the number of moves for the dominant strategy is greater than the rut percentage. If the rut percentage is less than 50%, then the dominant strategy percentage of moves will always be greater by definition.

### Setup of runs executed
In all runs, Figure A graphs swerves to straight moves, while Figure B plots the distribution of payoffs. There are 100 players, playing 100 rounds.

```python
import numpy as np

from ziagent_game import play_game
from ziagent_game.agents import Agent
from ziagent_game.graphs import graph_moves, graph_payoffs

players = 100
rounds = 100

def run_and_graph(players, rounds, memory, rut_prcnt, g_type):
    ''' Function to run games & analysis for all game variations.'''
    play_game.run(players, rounds, memory, rut_prcnt)

    # Create a list of agents to evaluate
    agents = Agent.instances
    swerves = []
    straights = []
    payoffs = []
    for a in agents:
        swerves.append(a[2])
        straights.append(a[3])
        payoffs.append(a[4])
    
    print "Figure A: %s" % g_type
    print "Swerve Mean: %s" % round(np.mean(swerves), 1)
    print "Swerve Std: %s" % round(np.std(swerves), 2)
    print "Straight Mean: %s" % round(np.mean(straights), 1)
    print "Straight Std: %s" % round(np.std(straights), 2)
    graph_moves(swerves, straights)
    print "----------------------------"
    
    print "Figure B: %s" % g_type
    print "Mean: %s" % round(np.mean(payoffs), 1)
    print "Std: %s" % round(np.std(payoffs), 2)
    graph_payoffs(payoffs, rounds)
```

#### Control Group
The control group has no memory, so it can never be in a rut. The rut percentage has no effect. 

```python
memory = 0
rut_prcnt = 0
g_type = 'Control Group - Memory %s -- rut = %s' % (str(memory), str(rut_prcnt))
run_and_graph(players, rounds, memory, rut_prcnt, g_type)
```

## Analysis
The effect of the rut check was not what I had expected. When the agents had no memory, the rut check has no effect. However, when memory was active, depending on the rut percentage used in the rut check there was a continuum of outcomes.
When the memory was set to three games and rut control was turned off, meaning that agents could play the same moves, there was a power law distribution in the total payoff amounts. There were the haves and the have nots with a mean payoff of 90.20 and a standard deviation of 23.78. By adding in the rut check and setting it to 100%,  the total payoffs are only slightly less than before, however the standard deviation is much less. 

With respect to shorter memory, a rut percentage check of less than 100% works againsts the agents, while with a longer memory, it helps the agents.  The highest total payoffs were during runs where agents had a memory of three games or five games and the rut percentage check was set to none (0) or 100%. This is followed by runs where agents had a memory of ten and a rut percentage check of none, 99 or 98. However, if the rut percentage check is set to 100, the agents are worse off. 

In general, the larger the mean of the total payoff, the larger the standard deviation in the total payoff. This is interesting, because in many of the runs there were outliers in the total payoff distribution. By lowering the rut percentage threshold, the payoff distribution and standard deviation of the total payoffs smoothed out to create a continuation in the distribution. 

### Conclusion

Does the rut control help? In the current setup of the model, rut control is really only beneficial to runs where agents have a long term memory. With respect to short term memory, rut control throws off the runs very quickly.
