## Setup

For the purpose of this analysis, the follow code is used to set up environment for the analysis. We start by importing standard libraries.


```python
import numpy as np
import pylab

import zi
```

Next, we setup our market size and the number of iterations that are market will go through. The market will automatically exit if no more trades are possible.

```python
market_size = 100
iterations = 1000
```

For the analysis, we create a few functions to handle calculations of averages and standard deviations for the results from each of the bulk runs. Lastly, there is a function that is used to display the distribution of activations.

```python
def calculate_avg_and_std(data, pos=None):
    if pos:
        values = [n[pos] for n in data]
    else:
        values = data
    avg = round(np.mean(values), 2)
    std = round(np.std(values, ddof=1), 2)
    
    pylab.hist(values, bins=40)
    pylab.show()
    return avg, std

def calculate_summary(data):
    r = {}
    print "AVERAGE TRADES"
    r['avg_trades'], r['std_avg_trades'] = calculate_avg_and_std(data, 2)
    print "AVERAGE SALES"
    r['avg_sale'], r['std_avg_sale'] = calculate_avg_and_std(data, 3)
    print "AVERAGE STD"
    r['avg_std'], r['std_avg_std'] = calculate_avg_and_std(data, 4)
    return r

def calculate_activations(data):
    b_values = []
    s_values = []
    for i in data:
        for v in i[1]:
            if v[0] == 's':
                s_values.append(v[1])
            else:
                b_values.append(v[1])
    
    print "DISTRIBUTION OF SELLER ACTIVATIONS:"
    pylab.hist(s_values, bins=40)
    pylab.show()
    
    print "DISTRIBUTION OF BUYER ACTIVATIONS:"
    pylab.hist(b_values, bins=40)
    pylab.show()
```

Then we setup a function to run the markets many times. We use 1000 as a default.

```python
# Run a market type 'n' number of times.
bulk_runs = 100
def run_market_n_times(func, args, n=bulk_runs):
    num_markets = n
    all_results = []
    for i in range(1,num_markets):
        results, activations = func(*args)
        all_results.append((results, activations))
    return all_results
```

## Application & Results

### Random Activation Run

**How agents are “random” activated in this model:**

With each turn a seller and a buyer agent are selected at random. If the seller and buyer are able to make a trade, they do so and both agents leave the market. If they are not able to make a trade, both the buyer and seller go back into the market and try again. 


#### Individual Run:

```python
results, activations = zi.run(market_size, iterations, True, save_plot=False, activation='random')
```

#### After 1000 runs:

```python
zi_args = (market_size, iterations, False, False, 'random')
random_act_results = run_market_n_times(zi.run, zi_args)
```

```python
calculate_activations(random_act_results)
```

```python
#Process and display stats from runs
calculations = calculate_summary([i[0] for i in random_act_results])
print "RESULTS"

for k,v in calculations.iteritems():
    print ' : '.join((k,str(v)))
```

### Uniform Activation Run

**How agents are “uniformly” activated in this model:**

With each iteration the agents shuffle themselves. While a full shuffle is unnecessary, it was applied, because the ease of application and the dataset was small enough to handle a full shuffle. 

After the shuffle, each seller looks at all the buyers to find a suitable match. Once the seller finds a buyer, the seller quits, because they only have one item in their inventory. The buyer quits also for the same reason. 

#### Individual run:

```python
results, activations = zi.run(market_size, iterations, True, save_plot=False, activation='uniform')
```

```python
# Running the market a bunch of times
zi_args = (market_size, iterations, False, False, 'uniform')
uniform_act_results = run_market_n_times(zi.run, zi_args)
```

#### After 1000 runs:

```python
calculate_activations(uniform_act_results)
```

```python
#Process and display stats from runs
calculations = calculate_summary([i[0] for i in uniform_act_results])
print "RESULTS"

for k,v in calculations.iteritems():
    print ' : '.join((k,str(v)))
```

## Analysis

In a comparison between the zero-intelligence model being set up with agents that are randomly activated versus uniformly activated, there are some interesting results. The uniform market was a more fruitful market for agents on average. There was a higher average sale price and in every case of the uniform activation, there were only about two agents that did not trade. 

With respect to the random market, it should be noted that buyers and sellers either found a match immediately or circulated around quite a bit. A few more found matches before the end of the iterations or no more matches were possible. 

With respect to the uniform market, it should be noted that the sellers are driving the market, so their distributions of activations isn’t as interesting as the buyers’ distribution of activations.
